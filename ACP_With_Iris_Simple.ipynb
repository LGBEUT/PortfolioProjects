```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

# Load data
data = load_iris()
X = data.data
y = data.target
target_names = data.target_names

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA transformation
pca = PCA(n_components=2)  # reduce dimensions to 2
X_pca = pca.fit_transform(X_scaled)

# Visualizing the PCA result
plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
# Charger les donnes Iris
iris = load_iris()
X = iris.data
y = iris.target 
target_name = iris.target_names
```

```python
print(f" X shape : {X.shape}, y shape : {y.shape}, target name : {target_name}")
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X) 

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca.shape
```

```python
X_pca
```

```python
pca.components_
```

```python
coeff = np.transpose(pca.components_[0:2, :])
```

```python
coeff
```

```python
n = coeff.shape[0]
```

```python
n
```

```python
# Crer la figure 
fig, ax = plt.subplots(figsize=(12,12))
# Vizu les catgorie avec les diffrentes couleurs
colors = ['r','g','b']
lw = 2

for colors, i, target_name in zip(colors, [0,1,2], target_names):
    plt.scatter(X_pca[y==i, 0], X_pca[y==i, 1], color=colors, label=target_name, lw=lw, alpha=0.8)


for i in range(n):                
    plt.arrow(0, 0, coeff[i, 0], coeff[i, 1], color='k', alpha=0.9, head_width=0.02, linestyle='--')
    plt.text(coeff[i, 0] * 1.15, coeff[i, 1] * 1.15, iris.feature_names[i], color='k', ha='center', va='center')

# Placer le cercle unitaire
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='--')
plt.gac().add_artist(circle)
```
